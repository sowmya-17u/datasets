```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
data = pd.read_csv('/food_coded.csv')

# Analyze the dataset
print(data.head())
print(data.info())
print(data.describe())

# Identify and handle missing values
print(data.isnull().sum())

# Fill missing values with mean
data['exercise'].fillna(data['exercise'].mean(), inplace=True)

# Fill missing values with median
data['exercise'].fillna(data['exercise'].median(), inplace=True)

# Fill missing values with mode
mode_values = data.mode().iloc[0]
data.fillna(mode_values, inplace=True)

# Verify the results
print(data.isnull().sum())
```
